```python
df= pd.read_html("https://en.wikipedia.org/wiki/List_of_Indian_states_and_union_territories_by_GDP")
```

```python
nominal_gdp = df[1]
gdp_by_year= df[3]
gdp_by_year_old= df[4]
nominal_gdp_growth_rate = df[5]
```

```python
nominal_gdp.rename(columns={"Data year[3][4][5][inconsistent]":"data_year" }, inplace=True)
```

```python
nominal_gdp_1= nominal_gdp
nominal_gdp_1['gdp_usd_billions']=nominal_gdp_1['Nominal GDP₹ lakh crore = INR trillions;USD billions']
```

```python
nominal_gdp_1['gdp_nominal_lakh_crore']= nominal_gdp_1['Nominal GDP₹ lakh crore = INR trillions;USD billions'].apply(lambda x: x.split(" ")[0])
nominal_gdp_1['gdp_nominal_lakh_crore']= nominal_gdp_1['gdp_nominal_lakh_crore'].apply(lambda x: x.split("₹")[1])
nominal_gdp_1['gdp_usd_billions']= nominal_gdp_1['gdp_usd_billions'].apply(lambda x: x.split("$")[1])
nominal_gdp_1['gdp_usd_billions']= nominal_gdp_1['gdp_usd_billions'].apply(lambda x: x.split(' ')[0])
```

```python
nominal_gdp_1.drop(columns=['Nominal GDP₹ lakh crore = INR trillions;USD billions'], inplace=True)
```

```python
nominal_gdp_1["data_year"] = nominal_gdp_1["data_year"].apply(lambda x: x.split("–")[0])
nominal_gdp_1["data_year"] = nominal_gdp_1["data_year"].apply(lambda x: x.split("-")[0])
nominal_gdp_1["gdp_usd_billions"] = nominal_gdp_1["gdp_usd_billions"].apply(lambda x: x.split(")")[0])
nominal_gdp_1["gdp_usd_billions"] = nominal_gdp_1["gdp_usd_billions"].apply(lambda x: x.split("billion")[0])
nominal_gdp_1
```

# end of processing for nominal gdp_per_year_1, this is the final product
 preprocessing gdp by year

```python
gdp_by_year = gdp_by_year.merge(gdp_by_year_old, on=['State/union territory'])
gdp_by_year.columns= gdp_by_year.columns.astype(str)
```

```python
gdp_by_year.rename(columns = {"2011–12 in ₹ crore":"2011",
                            "2012–13 in ₹ crore":"2012",
                            "2013–14 in ₹ crore":"2013",
                            "2014–15 in ₹ crore":"2014",
                           "2015–16 in ₹ crore":"2015",
                           "2016–17 in ₹ crore":"2016",
                           "2017–18 in ₹ crore":"2017",
                           "2018–19 in ₹ crore":"2018",
                           "2019–20 in ₹ crore":"2019",
                           "2020–21 in ₹ crore":"2020",
                              '2001–02':'2001',
                              '2002–03':'2002',
                              '2003–04':'2003',
                              '2004–05':'2004',
                              '2005–06':'2006',
                              '2006–07':"2007",
                              '2007–08':'2008',
                              '2008–09':'2009',
                              '2009–10':'2010'
                             }, inplace=True)
gdp_by_year.columns
gdp_by_year_lakh_crore= gdp_by_year
```

```python
nominal_gdp_growth = nominal_gdp_growth_rate
nominal_gdp_growth.set_index("State/union territory", inplace=True)
#make sure to comment this out if not running it the first time
nominal_gdp_growth.rename(columns={'2010–11':"2010",
                                        '2011–12':"2011",
                                        '2012–13':"2012",
                                       '2013–14': '2013',
                                       '2014–15':'2014',
                                        '2015–16':'2015',
                                        '2016–17': '2016',
                                        '2017–18': '2017',
                                        '2018-19':'2018',
                                        },
                         inplace=True)
```

```python
nominal_gdp_growth = nominal_gdp_growth.apply(lambda x: x.replace("–","x"))
```

```python
nominal_gdp_growth= nominal_gdp_growth.astype(str)
nominal_gdp_growth["2012"][6:7] = "-10"
nominal_gdp_growth["2013"][6:7] = "-5.8"
nominal_gdp_growth["2015"][10:11] = "19.10"
nominal_gdp_growth["2015"][11:12] = "-5.5" 

nominal_gdp_growth_for_absolute = nominal_gdp_growth

nominal_gdp_growth['2017'] = nominal_gdp_growth['2017'].replace("x", "10")
nominal_gdp_growth['2018'] = nominal_gdp_growth['2018'].replace("x", "11.2")

for columns in nominal_gdp_growth:
    nominal_gdp_growth[columns]= pd.to_numeric(nominal_gdp_growth[columns])

nominal_gdp_growth = nominal_gdp_growth.astype(float)

#same processing for the absolute dataframe


#nominal_gdp_growth_for_absolute['2017'] = nominal_gdp_growth_for_absolute['2017'].replace("x", "0")
#nominal_gdp_growth_for_absolute['2018'] = nominal_gdp_growth_for_absolute['2018'].replace("x", "0")

#for columns in nominal_gdp_growth_for_absolute:
#    nominal_gdp_growth_for_absolute[columns]= pd.to_numeric(nominal_gdp_growth_for_absolute[columns])
```

```python
# squeeze converts the first row in a numpy array and then we can subtract that from the whole dataframe 
nominal_relative_gdp_growth = nominal_gdp_growth - nominal_gdp_growth.loc['India'].squeeze()
```

```python
grd= pd.merge(pd.DataFrame(nominal_gdp_growth_for_absolute.reset_index()),
                              pd.DataFrame(nominal_relative_gdp_growth.reset_index()), 
                              on=['State/union territory'],
                             suffixes=["_a","_r" ])
grd.rename(columns ={"State/union territory": 'States'}, inplace=True)
```
